```python
# Extracting variables
X = df.iloc[:,[0,1,2,3]].values
# print(X)
```

```python
y = df.iloc[:,4].values
# print(y)
```

## Building classifier

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
claf = DecisionTreeClassifier()
claf.fit(X_train,y_train)
print("Trining Complete")
```

```python
# displaying text presentation of classifier
from sklearn import tree
txt_rep = tree.export_text(claf)
print(txt_rep)
```

## Tree visualization

```python
# visualizing the tree 
fig = plt.figure(figsize=(30,25))
fig = tree.plot_tree(claf, feature_names=X, class_names=y, filled=True, rounded=True)
```

# Give the new value to the model(test the model

```python
y_pred = claf.predict(X_test)
```

```python
a = [4,3,1,1]
test = n.array([a])
prediction = claf.predict(test)
print(prediction)
```

```python
b = [7,3,5,0]
test2 = n.array([b])
pred = claf.predict(test2)
print(pred)
```

```python
c = [6.5, 3.5, 5, 2.5]
test3 = n.array([c])
pred = claf.predict(test3)
print(pred)
```

# Evaluate Model

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test,y_pred))
```
