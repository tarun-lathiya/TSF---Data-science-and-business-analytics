```python
# installing all modules which required for the task
import numpy as n
import pandas as p
import matplotlib.pyplot as plt
import seaborn as sb
```

## Load data

```python
df = p.read_csv('Iris.csv',index_col='Id')
```

## Examine the data

```python
df.head()
```

```python
df.shape
```

```python
# Extracting variables
X = df.iloc[:,[1,2,3,4]].values
print(X)
```

```python
y = df.iloc[:,5].values
print(y)
```

## Data visualization

```python
sb.barplot(x=df['Species'],y=df['PetalWidthCm'],data=df)
```

```python
sb.pairplot(df)
```

## Building classifier

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
claf = DecisionTreeClassifier()
claf.fit(X_train,y_train)
print("Trining Complete")
```

```python
# displaying text presentation of classifier
from sklearn import tree
txt_rep = tree.export_text(claf)
print(txt_rep)
```

## Tree visualization

```python
# visualizing the tree 
fig = plt.figure(figsize=(25,20))
fig = tree.plot_tree(claf, feature_names=X, class_names=y, filled=True, rounded=True)
```

# Test the model 

```python
y_pred = claf.predict(X_test)
```
